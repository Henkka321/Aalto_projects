# Python montecarlo for fulll collection

```python
import numpy as np
import math
rng = np.random.default_rng() # numpy standard random sampling tool usable for statistical testing
def testRandomCollection(collection_size=100,duplicates_per_item= 5,duplicates_enabled = True):
    #simulates a collection process of drawing a random item 
    #collection_size: how many items are to be obtained
    #duplicates_per_item: optionally give a certain item if enough duplicates are drawn 
    

    items = np.zeros(collection_size)
    duplicates = 0
    tries = 0
    not_ready = True
    while not_ready:
        tries +=1
        sampled_item = rng.integers(low = 0, high = collection_size) # item number
        #check if new item
        if (items[sampled_item] == 1): #already collected
            if (duplicates_enabled):
                duplicates +=1
        else: #set collection item as obtained
                items[sampled_item] = 1
        #check if ready
        possible_items_from_duplicates = math.floor(duplicates/duplicates_per_item)
        if(np.sum(items) >= (collection_size-possible_items_from_duplicates)):
            not_ready= False
        
    return tries


print(testRandomCollection(collection_size=3,duplicates_per_item= 1,duplicates_enabled = True))

    


    



```

```python
# tests

# duplicates per item at 1 should always return collection size 

assert(testRandomCollection(collection_size=3,duplicates_per_item= 1,duplicates_enabled = True)== 3)
```

```python
# Run Montecarlo sampling to get expected number of trials

# test parameters
testsN = 1000
results = np.zeros(testsN)
collectionSize = 16
duplicatesPerItem = 5

for i in range( testsN):
    results[i] =testRandomCollection(collectionSize,duplicatesPerItem)
```

```python
import matplotlib.pyplot as plt
fix,ax = plt.subplots()
ax.hist(results, bins = 10)

plt.show()
```
